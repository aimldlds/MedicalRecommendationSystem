```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_predict, train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report
import pandas as pd

# ==============================
# 1. Encode target and split data
# ==============================
le = LabelEncoder()
df_work['prognosis_encoded'] = le.fit_transform(df_work['prognosis'])

X = df_work.drop(['prognosis', 'prognosis_encoded'], axis=1)
y = df_work['prognosis_encoded']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# ==============================
# 2. Define models
# ==============================
models = {
    'SVC': SVC(probability=True, random_state=42),
    'KNN': KNeighborsClassifier(),
    'Multinomial NB': MultinomialNB(),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42)
}


# ==============================
# 3. Evaluate with Stratified K-Fold CV
# ==============================
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = []

for name, model in models.items():
    y_pred_cv = cross_val_predict(model, X_train, y_train, cv=cv)  # predictions via CV

    acc = accuracy_score(y_train, y_pred_cv)
    precision, recall, f1, _ = precision_recall_fscore_support(
        y_train, y_pred_cv, average='macro', zero_division=0
    )

    results.append({
        "Model": name,
        "CV Accuracy": round(acc, 3),
        "Macro Precision": round(precision, 3),
        "Macro Recall": round(recall, 3),
        "Macro F1": round(f1, 3)
    })

# Create comparison table
results_df = pd.DataFrame(results).sort_values(by="Macro F1", ascending=False)
print("\n=== Model Comparison on Noisy Data (CV Metrics) ===")
print(results_df)

# ==============================
# 4. Final test set evaluation for the best model
# ==============================
best_model_name = results_df.iloc[0]["Model"]
best_model = models[best_model_name]
best_model.fit(X_train, y_train)
y_pred_test = best_model.predict(X_test)

print(f"\n=== Classification Report on Test Set for Best Model ({best_model_name}) ===")
print(classification_report(y_test, y_pred_test, target_names=le.classes_, zero_division=0))
```

```python
import shap
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np

# Prepare features and target from your noisy dataframe
X = df_work.drop(['prognosis', 'prognosis_encoded'], axis=1)
y = df_work['prognosis_encoded']

# Split your noisy data
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Train SVC model (ensure probability=True for predict_proba!)
model = SVC(kernel='linear', probability=True)
model.fit(X_train, y_train)

# Initialize SHAP KernelExplainer for SVC (model-agnostic)
background = X_train.sample(100, random_state=42)
explainer = shap.KernelExplainer(model.predict_proba, background)

# Compute SHAP values for a sample (limit to 100 rows for speed)
sample_data = X_train.sample(100, random_state=42)
shap_values = explainer.shap_values(sample_data, nsamples=100)

# Compute mean absolute SHAP values across all classes
mean_abs_shap_values = np.mean([np.abs(shp).mean(axis=0) for shp in shap_values], axis=0)

# Feature names
features = X_train.columns

# Rank features by importance
feature_importance = sorted(zip(features, mean_abs_shap_values), key=lambda x: x[1], reverse=True)

# Take top 10 features
top_features, importance = zip(*feature_importance[:10])

# Plot top 10 feature importances
plt.figure(figsize=(10,6))
plt.barh(top_features[::-1], importance[::-1], color='royalblue')
plt.xlabel('Mean Absolute SHAP Value')
plt.title('Top 10 Feature Importances (SHAP) - SVC')
plt.tight_layout()
plt.show()

# SHAP summary plot for the sample
shap.summary_plot(shap_values, sample_data)
```
